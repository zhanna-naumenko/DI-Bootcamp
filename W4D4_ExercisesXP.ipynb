```python
import pandas as pd
train_data = pd.read_csv('train.csv')
print(train_data.head(5))
```

```python
train_data = train_data.drop_duplicates()
print(train_data.duplicated().sum())
```

## **Exercise 2**

```python
missing_values = train_data.isnull()
print(missing_values.head())

count_missing = train_data.isnull().sum()
print(count_missing)
```

```python
clean_train_data = train_data.drop(labels='Cabin', axis=1)

clean_train_data.head(30)
```

```python
clean_train_data = clean_train_data.fillna(clean_train_data['Age'].mean())
clean_train_data.head(30)
```

```python
missing_values = clean_train_data.isnull()
print(missing_values.head())

count_missing = clean_train_data.isnull().sum()
print(count_missing)
```

## **Exercise 3**

```python
import pandas as pd
family_size_data = pd.DataFrame()
family_size_data['Family Size'] = clean_train_data['SibSp'] + clean_train_data['Parch']
family_size_data['Title'] = clean_train_data['Name']
print(family_size_data[['Title', 'Family Size']])
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
clean_train_data['Encode for Sex'] = label_encoder.fit_transform(clean_train_data['Sex'])
print(clean_train_data.head())
```

## **Exercise 4**

```python
# detect outliers for 'Fare' IQR
Q1 = clean_train_data['Fare'].quantile(0.25)
Q3 = clean_train_data['Fare'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers_fare = clean_train_data[(clean_train_data['Fare'] < lower_bound) | (clean_train_data['Fare'] > upper_bound)]
print(outliers_fare)
```

```python
# There is no need to deal with outlier of the column 'Age', as 70 and above year's old it is not actually outliers.
# This could be considered an outlier if the values were more than +120, then they would need to be changed.
# Concerning the outliers of the column 'Fare' this is also no longer an outlier, but a category of more expensive tickets according to z-score
# realistically, I wouldn’t change this data, but for practice I’ll change 'Fare'
#
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
new_clean_train_data = clean_train_data[clean_train_data['Fare'] < upper_bound]
new_clean_train_data.shape
plt.figure(figsize=(16,5))
plt.subplot(1,2,1)
sns.distplot(clean_train_data['Fare'])
plt.subplot(1,2,2)
sns.distplot(new_clean_train_data['Fare'])
plt.show()
```

```python
# detect outliers for 'Fare' z-score
from scipy import stats
z_scores = stats.zscore(clean_train_data['Fare'])
outliers_fare = clean_train_data[(z_scores > 3) | (z_scores < -3)]
print(outliers_fare.head(20))
```

```python
# detect outliers for 'Age' z-score
from scipy import stats
z_scores = stats.zscore(clean_train_data['Age'])
outliers_age = clean_train_data[(z_scores > 3) | (z_scores < -3)]
print(outliers_age.count())
```

```python
# detect outliers for 'Age' IQR
Q1 = clean_train_data['Age'].quantile(0.25)
Q3 = clean_train_data['Age'].quantile(0.75)
IQR = Q3 - Q1
print(f'Q1 Age: {Q1}')
print(f'Q3 Age: {Q3}')
print(f'IQR Age: {IQR}')
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(f'lower_bound Age: {lower_bound}')
print(f'upper_bound Age: {upper_bound}')
outliers_age2 = clean_train_data[(clean_train_data['Age'] < lower_bound) | (clean_train_data['Age'] > upper_bound)]
print(outliers_age2.count())
```

## **Exercise 5**

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# standardization for 'Age' and 'Fare'
standard_scaler = StandardScaler()
clean_train_data[['Age', 'Fare']] = standard_scaler.fit_transform(clean_train_data[['Age', 'Fare']])

# normalization for 'Age' and 'Fare'
min_max_scaler = MinMaxScaler()
clean_train_data[['Age', 'Fare']] = min_max_scaler.fit_transform(clean_train_data[['Age', 'Fare']])

print(clean_train_data)

numerical_columns = ['Age', 'Fare']
for column in numerical_columns:
    print(f"Column: {column}")
    print("Description:")
    print(clean_train_data[column].describe())
    print("Skewness:", clean_train_data[column].skew())
    print("Kurtosis:", clean_train_data[column].kurtosis())
    print("\n")
```

## **Exercise 7**

```python
bins = [0, 18, 30, 50, 70, 100]
labels = ['0-17', '18-29', '30-49', '50-69', '70+']
clean_train_data['AgeGroup'] = pd.cut(clean_train_data['Age'], bins=bins, labels=labels, right=False)
age_group_encoded = pd.get_dummies(clean_train_data['AgeGroup'], prefix='AgeGroup')
clean_train_data = pd.concat([clean_train_data, age_group_encoded], axis=1)
print(clean_train_data.head())
```
