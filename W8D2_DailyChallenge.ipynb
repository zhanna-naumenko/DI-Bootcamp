```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('words')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
```

```python
import re
import requests
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

test1 = "https://www.gutenberg.org/cache/epub/11/pg11.txt"
text2 = "https://www.gutenberg.org/cache/epub/12/pg12.txt"
test3 = "https://www.gutenberg.org/cache/epub/29042/pg29042.txt"

def load_texts(url):
    corpus = []
    response = requests.get(url)
    text = response.text
    # Clean the text by removing punctuation, converting to lowercase, and removing stopwords
    cleaned_text = re.sub(r'\W+', ' ', text.lower())
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(cleaned_text)
    cleaned_text = ' '.join([word for word in word_tokens if word not in stop_words])
    # Append the cleaned text to the corpus
    corpus.append(cleaned_text)
    return corpus

cleaned_text1 = load_texts(test1)
cleaned_text2 = load_texts(text2)
cleaned_text3 = load_texts(test3)

print(f"First 200 characters of text1: {cleaned_text1[:200]}")
print(f"First 200 characters of text2: {cleaned_text2[:200]}")
print(f"First 200 characters of text3: {cleaned_text3[:200]}")
```

```python
from nltk.tokenize import word_tokenize

# Tokenize the text
tokenized_text1 = word_tokenize(cleaned_text1[0])
tokenized_text2 = word_tokenize(cleaned_text2[0])
tokenized_text3 = word_tokenize(cleaned_text3[0])

# Print the first 150 tokens of each book
print("First 150 tokens of text1:")
print(tokenized_text1[:150])

print("\nFirst 150 tokens of text2:")
print(tokenized_text2[:150])

print("\nFirst 150 tokens of text3:")
print(tokenized_text3[:150])
```

```python
from nltk.stem import PorterStemmer


def stem_tokens(tokens):
    stemmer = PorterStemmer()
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    return stemmed_tokens

# Stem the tokens
stemmed_tokens1 = stem_tokens(tokenized_text1)
stemmed_tokens2 = stem_tokens(tokenized_text2)
stemmed_tokens3 = stem_tokens(tokenized_text3)

# Print the first 50 stemmed tokens of each book
print("First 50 stemmed tokens of text1:")
print(stemmed_tokens1[:50])

print("\nFirst 50 stemmed tokens of text2:")
print(stemmed_tokens2[:50])

print("\nFirst 50 stemmed tokens of text3:")
print(stemmed_tokens3[:50])
```

```python
import spacy

# Load spaCy pre-trained model
nlp = spacy.load('en_core_web_sm')

# Process the text with spaCy
doc1 = nlp(cleaned_text1[0])
doc2 = nlp(cleaned_text2[0])
doc3 = nlp(cleaned_text3[0])

# Extract lemmatized tokens
lemmatized_tokens1 = [token.lemma_ for token in doc1]
lemmatized_tokens2 = [token.lemma_ for token in doc2]
lemmatized_tokens3 = [token.lemma_ for token in doc3]

# Print the first 50 lemmatized tokens of each book
print("First 50 lemmatized tokens of text1:")
print(lemmatized_tokens1[:50])

print("\nFirst 50 lemmatized tokens of text2:")
print(lemmatized_tokens2[:50])

print("\nFirst 50 lemmatized tokens of text3:")
print(lemmatized_tokens3[:50])
```

Stemming and Lemmatization are both NLP techniques. They almost do the same thing, but a little bit in a different ways. Stemming is removing suffixes, while the Lemmatization involves reducing words to their base form.

```python
# Function to identify POS tags of each text
def identify_pos_tags(text):
    # Tokenize the text into words
    words = word_tokenize(text)
    # Identify POS tags for the words
    pos_tags = nltk.pos_tag(words)
    return pos_tags

# Identify POS tags for each text
pos_tags_text1 = identify_pos_tags(cleaned_text1[0])
pos_tags_text2 = identify_pos_tags(cleaned_text2[0])
pos_tags_text3 = identify_pos_tags(cleaned_text3[0])

# Print POS tags for the first 10 words of each text
print("POS tags for text 1:")
print(pos_tags_text1[:10])

print("\nPOS tags for text 2:")
print(pos_tags_text2[:10])

print("\nPOS tags for text 3:")
print(pos_tags_text3[:10])
```

```python
# Function to identify entities in the text
def identify_entities(text):
    # Tokenize the text into words
    words = word_tokenize(text)
    # Tag the words with POS tags
    pos_tags = nltk.pos_tag(words)
    # Identify named entities using ne_chunk
    named_entities = nltk.ne_chunk(pos_tags)
    return named_entities

# Identify entities for each text
entities_text1 = identify_entities(cleaned_text1[0])
entities_text2 = identify_entities(cleaned_text2[0])
entities_text3 = identify_entities(cleaned_text3[0])

# Print the identified entities for each text
print("Entities in text 1:")
print(entities_text1)

print("\nEntities in text 2:")
print(entities_text2)

print("\nEntities in text 3:")
print(entities_text3)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

def generate_word_cloud(text, title):
    # Generate the word cloud
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

    # Plot the word cloud
    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(title, fontsize=20)
    plt.axis('off')
    plt.show()


# Generate word clouds for each book
generate_word_cloud(cleaned_text1[0], 'Word Cloud for Book 1')
generate_word_cloud(cleaned_text2[0], 'Word Cloud for Book 2')
generate_word_cloud(cleaned_text3[0], 'Word Cloud for Book 3')
```

```python
from collections import Counter

# Concatenate all the cleaned text data into a single string
all_texts = cleaned_text1[0] + ' ' + cleaned_text2[0] + ' ' + cleaned_text3[0]

# Tokenize the text into words
words = all_texts.split()

# Count the frequency of each word
word_counts = Counter(words)

# Sort the words based on their frequency
sorted_words = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)

# Select the top five most frequent words
top_five_words = sorted_words[:5]

print("Top five most frequent words:")
for word, count in top_five_words:
    print(f"{word}: {count} times")
```

```python
# Sort the words by their frequency in descending order
sorted_word_counts = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)

# Extract the top five most frequent words
top_five_words = sorted_word_counts[:5]

# Print the document number, index, and count for each of the top five words
print("Top five most frequent words:")
for word, count in top_five_words:
    index = words.index(word) + 1  # Index starts from 1
    print(f"Word: {word}, Document: 1, Index: {index}, Count: {count}")
```

```python
import matplotlib.pyplot as plt

# Extract the top five most frequent words and their frequencies
top_five_words = [word for word, count in sorted_word_counts[:5]]
top_five_counts = [count for word, count in sorted_word_counts[:5]]

# Plot the pie plot
plt.figure(figsize=(8, 8))
plt.pie(top_five_counts, labels=top_five_words, autopct='%1.1f%%', startangle=140)
plt.title('Five Most Frequent Words')
plt.axis('equal')
plt.show()
```
