```python
import pandas as pd

data_happiness = pd.read_csv('2019.csv')
data_happiness.head()
data_happiness.info()
print(data_happiness.isnull().sum())
```

```python
# relationship between ‘Social support’ and ‘Score’
import matplotlib.pyplot as plt
x = data_happiness['Social support']
y = data_happiness['Score']

plt.scatter(x, y)
plt.title('Relationship between Social Support in the countries and the level of happiness')
plt.xlabel('Level of Social Support')
plt.ylabel('Happiness Score')
plt.show()
```

```python
# subplot that compares ‘GDP per Capita’ and ‘Healthy Life Expectancy’ across different regions
fig, (ax1, ax2) = plt.subplots(2,1, figsize=(20, 15))

x = data_happiness['Country or region']
y1 = data_happiness['GDP per capita']
y2 = data_happiness['Healthy life expectancy']
ax1.bar(x, y1, color='g', width=2, edgecolor='black')
ax1.set_title('GDP per capita by countries')
ax1.set_xlabel('Country')
ax1.set_ylabel('GDP per capita')
ax2.plot(x, y2, color='y')
ax2.set_title('Healthy life expectanc by countries')
ax2.set_xlabel('Country')
ax2.set_ylabel('Healthy life expectancy')
ax1.tick_params(axis='x', rotation=90)
ax2.tick_params(axis='x', rotation=90)
plt.tight_layout()
plt.show()
```
