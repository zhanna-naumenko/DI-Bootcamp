```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data_iris = sns.load_dataset('iris')
print(data_iris)
print(data_iris.isnull().sum())
unique_cgpa_values = data_iris['species'].unique()
print(unique_cgpa_values)
```

```python
# Generate a histogram for petal lengths.
import numpy as np

# plt.hist(x=data_iris['petal_length'], hue="species")
x = [data_iris.sepal_length[data_iris.species==sp_name] for sp_name in ['setosa', 'versicolor', 'virginica']]
plt.hist(x, 8, density=True, stacked=True)
plt.title('Histogram of Petal Lengths')
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
# a pair plot to show pairwise relationships in the dataset

sns.pairplot(data=data_iris, hue='species')
plt.show()
```

```python
# a scatter plot of sepal length vs. sepal width, color-coded by species
import plotnine
import plotnine.data
from plotnine import *
(
ggplot(data_iris, aes(x='sepal_length', y='sepal_width', color='species')) +
    geom_point(shape='d', size=4) + labs(title='Sepal Length vs. Sepal Width',
    x='Sepal Length',
    y='Sepal Width',
    color='Species')
)
```

```python
# an interactive 3D scatter plot mapping sepal length, sepal width, and petal length, color-coded by species
import plotly.express as px

fig = px.scatter_3d(data_iris, x='sepal_length', y='sepal_width', z='petal_length',
              color='species', symbol='species')
fig.update_layout(
    title='3D Scatter Plot of Iris Dataset',
    scene=dict(
        xaxis_title='Sepal Length',
        yaxis_title='Sepal Width',
        zaxis_title='Petal Length'
    ),
    legend=dict(
        title='Species'
    )
)
fig.show()
```
