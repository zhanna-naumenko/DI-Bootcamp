```python
# Exercise 1
import numpy as np
import pandas as pd
import scipy.stats as stats
from scipy import stats
import matplotlib.pyplot as plt

house_sizes = np.array([50, 70, 80, 100, 120])
# print(house_sizes)
house_prices =np.array([150000, 200000, 210000, 250000, 280000])
# print(house_prices)

slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")

# predict the price of a house that is 90 square meters
predicted_price = (slope * 90 + intercept).round(0)
print(f"Predicted price for a 90-square-meter house: {predicted_price}")

plt.scatter(house_sizes, house_prices)
plt.plot(90, predicted_price, 'gd')
x_reg = np.linspace(min(house_sizes), max(house_sizes), 100)  # Generating x values for the regression line
y_reg = slope * x_reg + intercept  # Calculating y values for the regression line
plt.plot(x_reg, y_reg, color='red', label='Regression Line')
plt.show()
```

```python
# Exercise 2

fertilizer_1 = np. array([5, 6, 7, 6, 5])
fertilizer_2 = np. array([7, 8, 7, 9, 8])
fertilizer_3 = np. array([4, 5, 4, 3, 4])

f_value, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)
print(f"F-value: {f_value}")
print(f"P-value: {p_value}")

# P-value is greater than 0.05, so we can reject the null hypothesis and assume that the fertilizers have similar effects
```

```python
# Exercise 3
# Example: Binomial Distribution
n, p = 10, 0.5
binom_dist = stats.binom(n, p)
print("PMF at 5 successes:", binom_dist.pmf(5))
```

```python
# Exercise 4
from scipy.stats import pearsonr
from scipy.stats import spearmanr
import matplotlib.pyplot as plt

data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})

# calculate Pearson Correlation between age and income
corr, _ = pearsonr(data['age'], data['income'])
print(f'Pearson Correlation Coefficient: {corr:.2f}')

# calculate Spearman Rank Correlation between age and income
corr, _ = spearmanr(data['age'], data['income'])
print(f'Spearman Rank Correlation: {corr:.2f}')
```

```python
# Exercise 5
import matplotlib.pyplot as plt
plt.scatter(data['age'], data['income'])
plt.title('Income vs. Age Frequency')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```
