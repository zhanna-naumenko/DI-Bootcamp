```python
import numpy as np
import random
import pandas as pd
#array from -5 to 35 with shape 10/12
temp_towns = np.random.randint(-5, 35, size=(10, 12))
print(temp_towns)
maen_temp_towns = temp_towns.mean()
print(maen_temp_towns.round(0))


towns=['Acrine', 'Blencogo', 'Cherrytown', 'Kinecardine', 'Saker Keep', 'Queenstown', 'Howe', 'Aelinmiley', 'Aermagh', 'Kelna']
data_temp = pd.DataFrame(temp_towns, index=range(len(temp_towns)), columns=['January', 'February', 'March',
                                                                         'April', 'May', 'June', 'July',
                                                                         'August', 'September', 'October',
                                                                         'November', 'December'])

# Add the 'Towns' column
data_temp['Towns'] = towns

# Set 'Towns' column as index
data_temp.set_index('Towns', inplace=True)
data_temp['Average temperature'] = data_temp.mean(axis=1).round(0)

print(data_temp)

#calculating average annual temperature from all datas
average_annual_temperature = data_temp.mean().sum() / 12

#calculating average annual temperature by towns
avg_temp = data_temp.mean()
print(f"\nAverage temperature by towns: \n{avg_temp.round(0)}")
print(f"\nAnnual average temperature: \n{average_annual_temperature.round(0)}")

max_temp = data_temp.idxmax()
print(f"\nThe city with the highest temperature by months: \n{max_temp}")

town_max_temp = data_temp['Average temperature'].idxmax()
print(f"Town with highest average temperature: {town_max_temp}")

min_temp = data_temp.idxmin()
print(f"\nThe city with the lowest temperature by months: \n{min_temp}")
town_min_temp = data_temp['Average temperature'].idxmin()
print(f"Town with highest average temperature: {town_min_temp}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

data_temp['Average temperature'] = data_temp.mean(axis=1).round(0)
list_towns = ['Acrine', 'Blencogo', 'Cherrytown', 'Kinecardine', 'Saker Keep', 'Queenstown', 'Howe', 'Aelinmiley', 'Aermagh', 'Kelna']
sns.barplot(data=data_temp, y='Average temperature', x='Towns', order=list_towns, estimator='mean', palette='Set2', hue='Towns')
plt.title('Average Temperature by Towns')
plt.xlabel('Towns')
plt.ylabel('Temperature')
plt.xticks(rotation=45)
plt.show()
```

```python
order_mnth = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
sns.barplot(data=data_temp, order=order_mnth, estimator='mean', palette='Set2')
plt.title('Avarage Temperature by Months')
plt.xlabel('Months')
plt.ylabel('Temperature')
plt.xticks(rotation = 45)
plt.show()
```

Looking through the randome data we can see that the town with highest average temperature is Aermagh and with the lowest temperature is Acrine. The highest temperature by months was on Februart and the lowest in July.
