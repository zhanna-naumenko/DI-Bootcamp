```python
import pandas as pd

data_appl = pd.read_csv('AAPL, 1D 01012007-12072023.csv')
data_appl
```

```python
from datetime import datetime

data_appl['time'] = pd.to_datetime(data_appl['time'], unit='s')
data_appl['formatted_dates'] = data_appl['time'].dt.strftime('%Y/%m/%d')
print(data_appl)
```

**Volume**: Volume refers to the total number of shares of Apple stock that were traded on a given day.

**VWAP (Volume-Weighted Average Price)**: VWAP is the average price of Apple stock for a given trading day, weighted by the trading volume at each price level.

**Open Price**: The open price is the price at which the first trade of Apple stock was executed at the beginning of the trading day.

**Close Price**: The close price is the price at which the last trade of Apple stock was executed at the end of the trading day.

**High Price**: The high price is the highest price at which Apple stock traded during the trading day.

**Low Price**: The low price is the lowest price at which Apple stock traded during the trading day.

**Number of Transactions**: This column represents the total number of individual transactions (buy or sell orders) that occurred during the trading day.

```python
from scipy.stats import ttest_ind
from scipy import stats

# T-Test and ANOVA of volume
monthly_volume = data_appl['volume'].resample('M').count()

pivot_table_volume = monthly_volume.reset_index().pivot_table(values='volume', index=monthly_volume.index.month, columns=monthly_volume.index.year, aggfunc='sum')

# T-test between two years
t_statistic, p_value = ttest_ind(pivot_table_volume[2007], pivot_table_volume[2008])

print("T-statistic 2007/2008:", t_statistic)
print("P-value 2007/2008:", p_value)

#ANOVA yearly
f_value, p_value = stats.f_oneway(pivot_table_volume[2007], pivot_table_volume[2008], pivot_table_volume[2009], pivot_table_volume[2010], pivot_table_volume[2011], pivot_table_volume[2012], pivot_table_volume[2013], pivot_table_volume[2014], pivot_table_volume[2015], pivot_table_volume[2016], pivot_table_volume[2017], pivot_table_volume[2018], pivot_table_volume[2019], pivot_table_volume[2020], pivot_table_volume[2021], pivot_table_volume[2022])
print("F-value 2007-2022:", f_value)
print("P-value 2007-2022:", p_value)
```

```python
from scipy.stats import shapiro
alpha = 0.05
numeric_data_appl = data_appl.select_dtypes(include=['float64', 'int64'])
for column in numeric_data_appl.columns:
  print(column)
  statistic, p_value = stats.shapiro(data_appl[column])
  print("Statistics", statistic, "p-value", p_value)
  if p_value < alpha:
    print("The null hypothesis can be rejected")
  else:
    print("The null hypothesis cannot be rejected")
```

```python
from scipy.stats import shapiro
data_appl['daily_returns'] = data_appl['close'].pct_change()

# Drop the first row (NaN value due to calculation)
data_appl.dropna(inplace=True)

# Test for normality using Shapiro-Wilk test
statistic, p_value = stats.shapiro(data_appl['daily_returns'])

print("Shapiro-Wilk Test:")
print("Statistic:", statistic)
print("P-value:", p_value)

if p_value < 0.05:
    print("The daily returns distribution is not normal (reject the null hypothesis).")
else:
    print("The daily returns distribution is normal (fail to reject the null hypothesis).")
```
