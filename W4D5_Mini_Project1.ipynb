```python
import pandas as pd
mental_health = pd.read_csv('survey.csv')
print(mental_health.head())
print(mental_health.dtypes)
print(mental_health.describe())
print(mental_health.isnull().sum())
male_count = (mental_health['Gender'] == 'male').sum()
female_count = (mental_health['Gender'] == 'female').sum()
print(f'Male rows: {male_count}')
print(f'Female rows: {female_count}')
```

```python
# deleting the rows that may not correspond to the actual interview age
mental_health = mental_health[(mental_health['Age'] >= 18) & (mental_health['Age'] <= 100)]
print(mental_health)
```

```python
mental_health['Gender'] = mental_health['Gender'].replace(dict.fromkeys(['male', 'm', 'man', 'Male', 'M', 'Male-ish', 'maile', 'Malr', 'Guy (-ish) ^_^', 'Cis Man', 'Make', 'make', 'something kinda male?', 'Cis Male', 'Male (CIS)', 'cis male', 'Mal', 'Mail', 'Nale', 'nale', 'Msle', 'msle' ], 'male'))
mental_health['Gender'] = mental_health['Gender'].replace(dict.fromkeys(['female', 'f', 'F', 'Woman', 'woman', 'Female', 'femail', 'famale', 'Femake', 'femake', 'Famale',  'famele', 'Famele', 'femele', 'Femele', 'Female (cis)', 'cis-female/femme', 'Cis Female'], 'female'))
print(mental_health)
male_count = (mental_health['Gender'] == 'male').sum()
female_count = (mental_health['Gender'] == 'female').sum()
print(f'Male rows: {male_count}')
print(f'Female rows: {female_count}')
```

```python
mental_health['self_employed'] = mental_health['self_employed'].replace(dict.fromkeys(['Yes', 'NA'], 'Yes/NA'))
mental_health['self_employed'] = mental_health['self_employed'].fillna(value='Yes/NA')
print(mental_health.head())
# yes_na_count = (mental_health['self_employed'] == 'Yes/NA').sum()
# no_count = (mental_health['self_employed'] == 'No').sum()
# print(f'Yes/NA: {yes_na_count}')
# print(f'No: {no_count}')
```

```python
#since a large percentage of data is missing we can make an imputation.
mental_health.drop(columns=['comments'], inplace=True)
```

```python
# since the 'state' column is mandatory only for people from the USA,
# and we are considering data not only from the USA and will not perform the analysis based on states, we can proceed with imputation
mental_health.drop(columns=['state'], inplace=True)
```

```python
mode_value = mental_health['work_interfere'].mode()[0]
mental_health['work_interfere'].fillna(mode_value, inplace=True)
```

```python
mental_health_val = mental_health.copy()
print(mental_health_val)
```

```python
mapping = {'Yes': 1, 'No': 0}
mental_health_val['family_history'] = mental_health_val['family_history'].map(mapping)
mental_health_val['treatment'] = mental_health_val['treatment'].map(mapping)
mental_health_val['remote_work'] = mental_health_val['remote_work'].map(mapping)
mental_health_val['tech_company'] = mental_health_val['tech_company'].map(mapping)

mapping2 = {'Yes/NA': 1, 'No': 0}
mental_health_val['self_employed'] = mental_health_val['self_employed'].map(mapping2)

print(mental_health_val)
```

```python
range_mapping = {
    '1-5': 1,
    '6-25': 2,
    '26-100': 3,
    '100-500': 4,
    'More than 1000': 5
}
mental_health_val['range_num_employeea'] = mental_health_val['no_employees'].map(range_mapping)
print(mental_health_val.head())
```

```python
age_mental_health_distribution = mental_health_val.groupby(['Age', 'mental_health_consequence']).size().reset_index(name='count')
print(age_mental_health_distribution)
```

```python
gender_mental_health_frequency = mental_health_val.groupby(['Gender', 'mental_health_consequence']).size().reset_index(name='count')
print(gender_mental_health_frequency)
```

```python
country_mental_health_frequency = mental_health_val.groupby(['Country', 'mental_health_consequence']).size().reset_index(name='count')
highest_mental_health_country = country_mental_health_frequency[country_mental_health_frequency['mental_health_consequence'] == 'Yes'].sort_values(by='count', ascending=False).iloc[0]
lowest_mental_health_country = country_mental_health_frequency[country_mental_health_frequency['mental_health_consequence'] == 'Yes'].sort_values(by='count').iloc[0]

print("Country with the highest reported rate of mental health issues:", highest_mental_health_country['Country'])
print("Country with the lowest reported rate of mental health issues:", lowest_mental_health_country['Country'])
```
