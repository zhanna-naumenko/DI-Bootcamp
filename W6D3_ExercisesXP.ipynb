```python
# Exercise 2
import scipy.stats as stats
import numpy as np

data = [12, 15, 13, 12, 18, 20, 22, 21]

#mean
mean_data = stats.tmean(data)
print(f"Mean: {mean_data}\n")

#SciPy does not have a separate median function
median_data = np.median(data)
print(f"Median: {median_data}\n")

#variance
variance_data = stats.tvar(data)
print(f"Variance: {variance_data}\n")

#standard deviation
std_dev_data = stats.tstd(data)
print(f"Srandart deviation: {std_dev_data}\n")
```

```python
# Exercise 3
from scipy import stats
import matplotlib.pyplot as plt
from scipy.stats import norm

mean = 50
std_dev = 10

height_distribution = np.random.normal(mean, std_dev, 500)
# print(height_distribution)

plt.hist(height_distribution, bins=30, density=True, alpha=0.8, color='g')
plt.show()

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'k', linewidth=2)
plt.title('Normal Distribution')
plt.show()
```

```python
# Exercise 4
rand_data1 = np.random.normal(20, 100, 500)
# print(f"Set of Data 1: \n{rand_data1}")
rand_data2 = np.random.normal(10, 200, 500)
# print(f"Set of Data 1: \n{rand_data2}")

T_test = scipy.stats.ttest_ind(rand_data1, rand_data2, axis=0, equal_var=True, nan_policy='propagate', permutations=None, random_state=None, alternative='two-sided')
print(T_test)
```

```python
# Exercise 5
import pandas as pd
from scipy import stats

data_iris = pd.read_csv('iris_dataset.csv')
data_iris.head()
# data_iris.info()

print(data_iris.isnull().sum())
data_iris['sepal_width'].fillna(data_iris['sepal_width'].mean(), inplace=True)
data_iris['petal_width'].fillna(data_iris['petal_width'].mean(), inplace=True)
print(data_iris.isnull().sum())

stats_desc_sepal_length = stats.describe(data_iris['sepal_length'])
print(f"Descriptive statistics for column 'sepal_length': \n{stats_desc_sepal_length}")
print(f"Number of values of column 'sepal_length': {stats_desc_sepal_length[0]}")
print(f"Min of column 'sepal_length': {stats_desc_sepal_length[1][0]}")
print(f"Max of column 'sepal_length': {stats_desc_sepal_length[1][1]}")
print(f"Mean of column 'sepal_length': {stats_desc_sepal_length[2]}")
print(f"Variance of column 'sepal_length': {stats_desc_sepal_length[3]}")
print(f"Skewness of column 'sepal_length': {stats_desc_sepal_length[4]}")
print(f"Kurtosis of column 'sepal_length': {stats_desc_sepal_length[-1]}\n")

stats_desc_sepal_width = stats.describe(data_iris['sepal_width'])
print(f"Descriptive statistics for column 'sepal_width': \n{stats_desc_sepal_width}\n")

stats_desc_petal_length = stats.describe(data_iris['petal_length'])
print(f"Descriptive statistics for column 'petal_length': \n{stats_desc_petal_length}\n")

stats_desc_petal_width = stats.describe(data_iris['petal_width'])
print(f"Descriptive statistics for column 'petal_width': \n{stats_desc_petal_width}\n")
```
