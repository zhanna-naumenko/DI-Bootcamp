Type: Supervised Learning (Regression)

This problem involves predicting future numerical values based on historical data. A supervised learning regression model is most suitable because it learns patterns from labeled data (past prices) to forecast continuous outcomes like stock prices.


Organizing a Library of Books

This task requires grouping books based on similarities (e.g., themes, keywords, genres) without predefined labels. Clustering algorithms like K-Means can automatically organize books into meaningful categories.

**Exercise 5 : Designing an Evaluation Strategy for Different ML Models**

Supervised models are evaluated using metrics like accuracy, recall, and ROC-AUC with validation techniques such as cross-validation. Unsupervised models rely on internal validation metrics like silhouette score and domain interpretation. Reinforcement learning models are evaluated based on cumulative reward, convergence behavior, and policy efficiency, but are often harder to assess due to dynamic environments.
