Applications of Machine Learning

Machine Learning is widely used across industries:

Finance: Loan default prediction and fraud detection
Healthcare: Disease diagnosis and patient risk prediction
E-commerce: Personalized product recommendations.

Types of Machine Learning
- Supervised Learning

Definition: Learns from labeled data to predict outcomes
Example: Predicting whether a loan will default

- Unsupervised Learning

Definition: Finds patterns in unlabeled data
Example: Grouping customers into segments based on behavior

- Reinforcement Learning

Definition: Learns through interaction using rewards and penalties
Example: A robot learning to navigate a maze
