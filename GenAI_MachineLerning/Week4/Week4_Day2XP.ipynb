**Exercise 1 : Matrix Operations**

```python
# create matrix
matrix = np.array([
    [4, 2, 1],
    [0, 5, 3],
    [2, 1, 3]
])

# optional: convert to DataFrame for better visualization
df = pd.DataFrame(matrix)
print(df)

det = np.linalg.det(matrix)
print("Determinant:", det)

inv_matrix = np.linalg.inv(matrix)

# convert to DataFrame for readability
inv_df = pd.DataFrame(inv_matrix)

print("Inverse Matrix:")
print(inv_df)
```

**Exercise 6 : Basic Hypothesis Testing**

```python
import numpy as np

np.random.seed(42)

# before training
productivity_before = np.random.normal(loc=50, scale=10, size=30)

# after training
productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)
```

```python
mean_before = np.mean(productivity_before)
mean_after = np.mean(productivity_after)

print(f"Mean Before: {round(mean_before, 3)}")
print(f"Mean After: {round(mean_after, 3)}")
print(f"Difference: {round(mean_after - mean_before, 3)}")
```

```python
# differences
diff = productivity_after - productivity_before

# mean and std of differences
mean_diff = np.mean(diff)
std_diff = np.std(diff, ddof=1)

# sample size
n = len(diff)

# t-statistic
t_stat = mean_diff / (std_diff / np.sqrt(n))

print("t-statistic:", t_stat)
```

```python
from scipy.stats import ttest_rel

t_stat, p_value = ttest_rel(productivity_after, productivity_before)

print("t-stat:", t_stat)
print("p-value:", p_value)
```
