**1. Data Preparation**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# reproducibility
np.random.seed(42)

# generate synthetic temperature data (-5 to 35)
data = np.random.uniform(-5, 35, size=(10, 12))

# city names
cities = [
    "New York", "Los Angeles", "Chicago", "Houston", "Phoenix",
    "Philadelphia", "San Antonio", "San Diego", "Dallas", "San Jose"
]

# months
months = [
    "Jan", "Feb", "Mar", "Apr", "May", "Jun",
    "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"
]

# create DataFrame
temperature = pd.DataFrame(data, index=cities, columns=months)

# round for readability
temperature = temperature.round(2)

temperature.head()
```

**2. Data Analysis**

```python
# annual average temperature per city
temperature["Annual Avg"] = temperature.mean(axis=1)

# highest and lowest
hottest_city = temperature["Annual Avg"].idxmax()
coldest_city = temperature["Annual Avg"].idxmin()

print("Hottest city:", hottest_city)
print("Coldest city:", coldest_city)

temperature[["Annual Avg"]].sort_values(by="Annual Avg", ascending=False)
```

**3. Data Visualization**

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10,6))

sns.barplot(
    x=temperature["Annual Avg"],
    y=temperature.index
)

plt.title("Average Annual Temperature by City")
plt.xlabel("Temperature (°C)")
plt.ylabel("City")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.heatmap(temperature.iloc[:, :-1], cmap="coolwarm", annot=True, fmt=".1f")

plt.title("Monthly Temperature Distribution by City")
plt.xlabel("Month")
plt.ylabel("City")

plt.tight_layout()
plt.show()
```
