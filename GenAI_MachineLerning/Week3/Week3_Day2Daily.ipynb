```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
```

```python
salary = pd.read_csv('/content/datascience_salaries.csv')
```

```python
# Normalize salary
scaler = MinMaxScaler()
salary['salary_normalized'] = scaler.fit_transform(salary[['salary']])
print(salary[['salary', 'salary_normalized']].head())
```

```python
# Encode categorical variables
salary_encoded = pd.get_dummies(
    salary,
    columns=['job_title','job_type','experience_level','location','salary_currency'],
    drop_first=True
)
```

```python
# PCA dimensionality reduction
pca = PCA(n_components=2)
pca_features = pca.fit_transform(salary_encoded)
```

```python
# Salary statistics
salary_stats = salary.groupby('experience_level')['salary'].agg(['mean','median'])

print(salary_stats)
```
