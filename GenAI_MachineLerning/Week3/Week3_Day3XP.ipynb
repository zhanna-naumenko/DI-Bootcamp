**Exercise 2: Descriptive Statistics**

```python
data = [12, 15, 13, 12, 18, 20, 22, 21]
mean_np = np.mean(data)
print(f"Mean of data: {mean_np}")
median_np = np.median(data)
print(f"Median of data: {median_np}")
variance_np = np.var(data)
print(f"Variance of data: {variance_np}")
std_dev_np = np.std(data)
print(f"Standard Deviation of data: {std_dev_np}")
```

**Exercise 3: Understanding Distributions**

```python
# Parameters
mean = 50
std = 10

# Generate x values
x = np.linspace(mean - 4*std, mean + 4*std, 1000)

# Generate normal distribution values
y = norm.pdf(x, mean, std)

# Plot the distribution
plt.plot(x, y)
plt.title("Normal Distribution (mean=50, std=10)")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.show()
```

**Exercise 4: T-Test Application**

```python
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

# Calculate means
mean_data1 = np.mean(data1)
print(f"Mean of data1: {mean_data1}")
mean_data2 = np.mean(data2)
print(f"Mean of data: {mean_data2}")
```

```python
# Calculate standard deviation
std_data1 = np.std(data1, ddof=1)
print(f"Standard Deviation of data1: {std_data1}")
std_data2 = np.std(data2, ddof=1)
print(f"Standard Deviation of data2: {std_data2}")
```

```python
from scipy.stats import ttest_ind
# Perform independent T-test
t_stat, p_value = ttest_ind(data1, data2)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

**Exercise 5: Linear Regression Analysis**

```python
from sklearn.linear_model import LinearRegression

# Dataset
house_sizes = np.array([50, 70, 80, 100, 120]).reshape(-1, 1)
house_prices = np.array([150000, 200000, 210000, 250000, 280000])
```

```python
model = LinearRegression()
model.fit(house_sizes, house_prices)
```

```python
slope = model.coef_[0]
intercept = model.intercept_

print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
```

Slope: ≈ 1857.14

Intercept: ≈ 60714.29

Predicted price for 90 m²: ≈ 227,857

Interpretation: Each additional square meter increases the house price by about 1,857 currency units.

**Exercise 6: Understanding ANOVA**

```python
from scipy.stats import f_oneway

fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

f_value, p_value = f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

print(f"F-value: {f_value}")
print(f"P-value: {p_value}")
```

1.What are the F-value and P-value?
F-value ≈ 19.9
P-value ≈ 0.00018

2.Based on the P-value, do the fertilizers have significantly different effects on plant growth?
Yes, the fertilizers have significantly different effects, because P-value < 0.05
This means the differences between the groups are statistically significant, so at least one fertilizer produces different plant growth compared to the others.

3.Explain what would happen if the P-value were greater than 0.05.
If P-value > 0.05, then we fail to reject the null hypothesis. This would mean that there is no statistically significant difference between the fertilizers.
Any observed differences in plant growth are likely due to random variation rather than the fertilizer type.
In that case, we would conclude that all fertilizers perform similarly.
