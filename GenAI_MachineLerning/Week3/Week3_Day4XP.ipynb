**Exercise 1: Understanding Data Visualization**

**Explain why data visualization is important in data analysis.**

It is important because it helps transform complex data into clear and understandable visual formats such as charts, graphs, and dashboards. Visualizations make it easier to identify patterns, trends, and outliers that might not be obvious when looking at raw data. They also help analysts communicate insights effectively to stakeholders, enabling better and faster decision-making based on the data.

**Describe the purpose of a line graph in data visualization.**

The purpose of a line graph in data visualization is to show how a variable changes over time or across a continuous range. It helps identify trends, patterns, and fluctuations in the data, making it easier to see increases, decreases, or long-term developments.

**Exercise 2: Creating a Line Plot for Temperature Variation**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
temperature = [72, 74, 76, 80, 82, 78, 75]
days = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]

plt.plot(days, temperature, color = 'g', marker='o')
plt.title('Temperature Variation')
plt.xlabel('Days')
plt.ylabel('Temperature (°F)')
plt.show()
```

**Exercise 3: Visualizing Monthly Sales with a Bar Chart**

```python
sales = [5000, 5500, 6200, 7000, 7500]
months = ["January", "February", "March", "April", "May"]

plt.bar(months, sales, color = 'c')
plt.xlabel('Month')
plt.ylabel('Sales Amount ($)')
plt.title('Monthly Sales')
plt.show()
```

**Exercise 4: Visualizing the Distribution of CGPA**

```python
mental_health = pd.read_csv('/content/Student Mental health.csv')
mental_health.head()
```

```python
mental_health.info()
```

```python
mental_health.describe()
```

```python
mental_health.columns
```

```python
print(f"The number of duplicates: {mental_health.duplicated().sum()}")
```

```python
print(f"The number of missing values: {mental_health.isnull().sum().sum()}")
```

```python
mental_health.isnull().sum()
```

```python
mental_health["Age"] = mental_health["Age"].fillna(mental_health["Age"].median())
```

```python
mental_health.isnull().sum()
```

```python
print(mental_health['Marital status'].unique())
print(mental_health['Do you have Depression?'].unique())
print(mental_health['Do you have Anxiety?'].unique())
print(mental_health['Do you have Panic attack?'].unique())
print(mental_health['Did you seek any specialist for a treatment?'].unique())
```

```python
# Convert Yes/No columns to binary (1/0)
binary_cols = [
    'Marital status',
    'Do you have Depression?',
    'Do you have Anxiety?',
    'Do you have Panic attack?',
    'Did you seek any specialist for a treatment?'
]

mental_health[binary_cols] = mental_health[binary_cols].apply(lambda col: col.map({'Yes': 1, 'No': 0}))
```

```python
# Convert Timestamp to datetime
mental_health['Timestamp'] = pd.to_datetime(mental_health['Timestamp'], format='mixed', dayfirst=True)
```

```python
mental_health['Age'] = mental_health['Age'].astype(int)
```

```python
mental_health.head()
```

```python
mental_health['What is your CGPA?'] = mental_health['What is your CGPA?'].str.strip()
```

```python
print(mental_health['What is your CGPA?'].unique())
```

```python
sns.histplot(data=mental_health, x="What is your CGPA?", color='steelblue')

plt.title('Distribution of CGPA')
plt.xlabel('CGPA')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

**Exercise 5: Comparing Anxiety Levels Across Different Genders**

```python
sns.barplot(data=mental_health, x='Choose your gender', y='Do you have Anxiety?', palette='pastel')

plt.title('Anxiety Levels Across Different Genders')
plt.xlabel('Gender')
plt.ylabel('Proportion with Anxiety')
plt.show()
```

**Exercise 6: Exploring the Relationship Between Age and Panic Attacks**

```python
sns.scatterplot(data=mental_health, x='Age', y='Do you have Panic attack?', hue='Do you have Panic attack?', palette='pastel', style='Do you have Panic attack?', markers={0: 'o', 1: 'X'},
                s=100)

plt.title('Relationship Between Age and Panic Attacks')
plt.xlabel('Age')
plt.ylabel('Proportion with Panic Attacks')
plt.show()
```
