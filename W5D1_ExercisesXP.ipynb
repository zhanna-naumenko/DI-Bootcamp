## **Exercise 4**

```python
import pandas as pd
import numpy as np
url = 'https://raw.githubusercontent.com/intern2grow/sales-data-analysis/main/sales_data.csv'
sales_data = pd.read_csv(url)

# the total quantity of products sold by the company
totla_quan = sales_data.sum()['quantity']
print(f'Total quantity sold by Company: {totla_quan}')

# the category that had the highest revenue and calculate how much revenue it generated
reset_sales_data = sales_data.set_index(['category'])
best_category = reset_sales_data.groupby('category')['revenue'].sum().idxmax()
best_category_value = reset_sales_data.groupby('category')['revenue'].sum().max()
print(f'The best category is {best_category} with value {best_category_value}')

# the average revenue per sale in the dataset
avr_per_revenue = sales_data['revenue'].mean()
print(f'Average revenue: {avr_per_revenue}')

# the total revenue generated in each quarter of the year (Q1, Q2, Q3, Q4)
sales_data['date'] = pd.to_datetime(sales_data['date'])
quarterly_revenue = sales_data.groupby(sales_data['date'].dt.to_period('Q'))['revenue'].sum()
plt.figure(figsize=(10, 6))
quarterly_revenue.plot(kind='bar', color='g')
plt.title('Total revenue by quarters')
plt.xlabel('Quarter')
plt.ylabel('Total revenue')
plt.show()
```

```python
# Create a simple line plot using Matplotlib. Plot the function y = x^2 for x values ranging from -10 to 10.
# Add a title and label the x and y axes
x = [x for x in range(-10, 11, 1)]
y = [val**2 for val in x]
plt.plot(x, y)
plt.title('Plot of the function y = x^2')
plt.xlabel('x values')
plt.ylabel('y values')
plt.show()
```

```python
# Generate a bar chart to display the sales data of four different products (A, B, C, D) in a week.
# The sales values are 15, 30, 45, and 20 respectively. Label the chart appropriately.
import pandas as pd

sales = {'Product': ['A', 'B', 'C', 'D'],
         'Sales': [15, 30, 45, 20]}

sales_val = pd.DataFrame(sales)
print(sales_val)

x = sales_val['Product']
y = sales_val['Sales']
plt.bar(x, y, color=['g', 'b', 'r', 'y'])
plt.title('Total sales of the products by category')
plt.xlabel('Product')
plt.ylabel('Amount')
plt.show()
```

```python
# Create a pie chart representing the following data about favorite fruits of a group of people:
# Apples (40%), Bananas (30%), Cherries (20%), and Dates (10%).
# Customize the chart with different colors for each fruit and add a legend.

fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
favorite = [40, 30, 20, 10]
plt.figure(figsize=(8, 8))
plt.pie(favorite, labels=fruits, autopct='%1.1f%%')
plt.title('Preferable fruits')
plt.legend(labels=fruits, loc="best")
plt.show()
```
