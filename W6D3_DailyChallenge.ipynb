```python
airplane_crash['Date'] = pd.to_datetime(airplane_crash['Date'])
airplane_crash['Time'] = pd.to_datetime(airplane_crash['Time'], format='%H:%M', errors='coerce')
```

```python
airplane_crash['AC Type'].fillna('Other', inplace=True)
airplane_crash['Location'].fillna('Other', inplace=True)
airplane_crash['Operator'].fillna('Other', inplace=True)
airplane_crash = airplane_crash.drop('Flight #', axis=1)
airplane_crash['Route'].fillna('Other', inplace=True)
airplane_crash['Registration'].fillna('Other', inplace=True)
airplane_crash = airplane_crash.drop('cn/ln', axis=1)
print(airplane_crash.isnull().sum())
```

```python
# filling in the null values from column Abourd with values in column Fatalities, as somebody died, that means they was on the plane
airplane_crash['Aboard'].fillna(airplane_crash['Fatalities'], inplace=True)
```

```python
# filling in null values in column Aboard with the mean by AC type
mean_by_ac_type = airplane_crash.groupby('AC Type')['Aboard'].transform('mean')
airplane_crash['Aboard'].fillna(mean_by_ac_type, inplace=True)
```

```python
# filling in the null values of column Aboard with mean
airplane_crash['Aboard'].fillna(airplane_crash['Aboard'].mean(), inplace=True)
```

```python

airplane_crash['Fatalities'].fillna(airplane_crash['Aboard'], inplace=True)
```

```python
import numpy as np
mean_by_ac_type = airplane_crash.groupby('AC Type')['Aboard'].transform('mean')
airplane_crash['Aboard'] = np.where(airplane_crash['Aboard'] == 0, mean_by_ac_type, airplane_crash['Aboard'])
```

```python
print((airplane_crash['Aboard'] == 0).sum())
```

```python
mean_aboard = airplane_crash.loc[airplane_crash['Aboard'] != 0, 'Aboard'].mean()
airplane_crash['Aboard'] = airplane_crash['Aboard'].replace(0, mean_aboard)
```

```python
print((airplane_crash['Aboard'] == 0).sum())
```

```python
airplane_crash['Fatalities Passangers'].fillna(airplane_crash['Aboard Passangers'], inplace=True)
airplane_crash['Fatalities Crew'].fillna(airplane_crash['Aboard Crew'], inplace=True)
```

```python
airplane_crash['Aboard Passangers'].fillna(airplane_crash['Fatalities Passangers'], inplace=True)
airplane_crash['Aboard Crew'].fillna(airplane_crash['Fatalities Crew'], inplace=True)
```

```python
mean_by_ac_type = airplane_crash.groupby('AC Type')['Aboard Crew'].transform('mean')
airplane_crash['Aboard Crew'].fillna(mean_by_ac_type, inplace=True)
```

```python
airplane_crash['Aboard Passangers'] = airplane_crash['Aboard Passangers'].fillna(airplane_crash['Aboard'] - airplane_crash['Aboard Crew'])
```

```python
airplane_crash['Fatalities Passangers'].fillna(airplane_crash['Aboard Passangers'], inplace=True)
airplane_crash['Fatalities Crew'].fillna(airplane_crash['Aboard Crew'], inplace=True)
```

```python
airplane_crash.dropna(subset=['Aboard Crew'], inplace=True)
```

```python
airplane_crash = airplane_crash.drop('Time', axis=1)
airplane_crash = airplane_crash.drop('Ground', axis=1)
airplane_crash = airplane_crash.drop('Summary', axis=1)
```

```python
print(airplane_crash.isnull().sum())
```

```python
airplane_crash['Year'] = airplane_crash['Date'].dt.year
airplane_crash['Month'] = airplane_crash['Date'].dt.month
print(airplane_crash['Year'].unique())
print(airplane_crash['Month'].unique())
```

```python
import matplotlib.pyplot as plt

frequency_of_aircrash = airplane_crash['Year'].value_counts().sort_index()


plt.figure(figsize=(10, 6))
plt.bar(frequency_of_aircrash.index, frequency_of_aircrash.values, color='m')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.title('Frequency of Airplane Crashes by Year')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import calendar

frequency_of_aircrash_mnth = airplane_crash['Month'].value_counts().sort_index()

month_names = [calendar.month_name[i] for i in frequency_of_aircrash_mnth.index]

plt.figure(figsize=(10, 6))
plt.bar(month_names, frequency_of_aircrash_mnth.values, color='c')
plt.xlabel('Month')
plt.ylabel('Frequency')
plt.title('Frequency of Airplane Crashes by Months')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
operator_frequency = airplane_crash['Operator'].value_counts().head(10)

plt.figure(figsize=(10, 6))
operator_frequency.plot(kind='bar', color='green')
plt.xlabel('Operator')
plt.ylabel('Frequency')
plt.title('Top 10 Air crashes by Operators')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
ac_type_frequency = airplane_crash['AC Type'].value_counts().head(10)

plt.figure(figsize=(10, 6))
ac_type_frequency.plot(kind='bar', color='green')
plt.xlabel('AC Type')
plt.ylabel('Frequency')
plt.title('Top 10 Air crashes by AC Type')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
route_frequency = airplane_crash['Route'].value_counts().head(10)

plt.figure(figsize=(10, 6))
route_frequency .plot(kind='bar', color='green')
plt.xlabel('Route')
plt.ylabel('Frequency')
plt.title('Top 10 Air crashes by Route')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

fatalities_by_year = airplane_crash.groupby(airplane_crash['Year'])['Fatalities'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(data=fatalities_by_year, x='Year', y='Fatalities', marker='o', color='m')
plt.xlabel('Year')
plt.ylabel('Number of Fatalities')
plt.title('Number of Fatalities by Year')
plt.grid(True)
plt.tight_layout()
```

```python
fatalities_ac_type = airplane_crash.groupby(airplane_crash['AC Type'])['Fatalities'].sum().reset_index()

top_10_ac_types = fatalities_ac_type.nlargest(10, 'Fatalities')

plt.figure(figsize=(10, 6))
sns.lineplot(data=top_10_ac_types, x='AC Type', y='Fatalities', marker='o', color='r')
plt.xlabel('AC Type')
plt.ylabel('Number of Fatalities')
plt.title('Number of Fatalities by AC Type')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
```

```python
fatalities_route = airplane_crash.groupby(airplane_crash['Route'])['Fatalities'].sum().reset_index()

top_10_routes = fatalities_route.nlargest(10, 'Fatalities')

plt.figure(figsize=(10, 6))
sns.lineplot(data=top_10_routes, x='Route', y='Fatalities', marker='o', color='r')
plt.xlabel('Route')
plt.ylabel('Number of Fatalities')
plt.title('Number of Fatalities by Route')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
```

**Report**

We can observe a rise in the frequency of air crashes starting from 1920, with a significant decrease only noticeable from 2011. However, it is unclear whether this is a consequence of increased air traffic or other factors. Similarly, the substantial decrease in aircraft incidents from 2020 may be related to the reduction in flights due to the COVID-19 pandemic and subsequent bankruptcies of major airlines, which naturally affected flight numbers post-pandemic.

There are no significant differences in frequency between months, indicating that the timing of flights does not affect the likelihood of a crash. Even a slight increase in frequency in December cannot be directly attributed to weather conditions or other factors, as it may be due to heightened holiday travel during Christmas and New Year.

Regarding the top 10 air crash frequencies by operators, it is unclear whether this reflects the quality of planes, engines, etc., or if it is a result of measurement errors that do not consider the percentage of air crashes relative to the total number of flights operated by each airline. However, it is worth noting the notably high frequency of air crashes associated with the Aeroflot operator, as well as the Douglas DC-3 aircraft type, which also exhibits a high frequency of crashes.

Additionally, there appears to be a higher probability of air crashes during training, sightseeing, or test flights.
